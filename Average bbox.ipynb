```python
import pandas as pd
directory = "train.csv" ##change directory here
```

```python
df = pd.read_csv(directory)
df.head(10)
```

```python
del df['image_id']
```

```python
del df['width']
del df['height']
del df['source']
```

```python
df.head(10)
```

```python
df['bbox'] = df['bbox'].str[1:].str[:-1]
df.head(10)
```

```python
df[['xmin','ymin', 'width', 'height']] = df['bbox'].str.split(',',expand=True)
```

```python
df.head(10)
```

```python
del df['bbox']
```

```python
df['width'] = pd.to_numeric(df['width'],errors='coerce')
df['height'] = pd.to_numeric(df['height'],errors='coerce')
df.head(10)
```

```python
print(df["width"].mean())
print(df["height"].mean())
```
